```python
import pandas as pd
import matplotlib.pyplot as plt

# Load a CSV file into a DataFrame
df = pd.read_csv('world_population.csv')

# Display the first few rows
print(df.head())
print(df.info())


df1=df.head(40)

# Step 4: Plot the bar chart
plt.figure(figsize=(30, 10))
plt.bar(df1['CCA3'], df1['2015 Population'], color='skyblue')

plt.xlabel('Population')

plt.title('Top 20 Most Populous Countries in 2015',fontsize=30,color='red')
plt.gca().invert_xaxis()  # Largest population at the top
plt.tight_layout()
plt.show()
```

```python
plt.hist(df1['2022 Population'], bins=100, color='skyblue', edgecolor='black')
plt.xlabel('Population')

plt.title('Top 20 Most Populous Countries in 2015',fontsize=30,color='red')
plt.gca().invert_xaxis()  # Largest population at the top
plt.tight_layout()
plt.show()
```
