```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python



df = pd.read_csv("tested.csv")
print(df.head())
```

```python
df.head()           # View first 5 rows
df.tail()           # View last 5 rows
df.info()           # Data types and non-null info
```

```python
df.describe()       # Summary statistics for numeric columns
df['PassengerId'].value_counts()  # For categorical columns
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
male=df[df['Sex']=='male']
print(male)
female=df[df['Sex']=='female']
print(female)
```

```python

import matplotlib.pyplot as plt

plt.hist(male['Age'],color='blue')
plt.title("  Histogram for male ages")
plt.show()
```

```python
plt.hist(female['Age'],color='blue')
plt.title("  Histogram for female ages")
plt.show()
```

```python

plt.bar(df['Sex'],df['Age'],color='red')
plt.title("Titanic: Age vs Sex -- Bar graph")
plt.show()
```

```python
# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Fare'], color='red', label='male', alpha=0.6)
plt.scatter(df['Age'], df['Fare'], color='green', label='female', alpha=0.6)

plt.title('Titanic: Age vs Fare (Colored by Sex) -- scatter plot' )
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(df['Age'], df['Fare'], color='red', label='male', alpha=0.6)
plt.plot(df['Age'], df['Fare'], color='orange', label='female', alpha=0.6)
plt.title("Titanic: Age vs Fare (Colored by sex) -- linear graph")
plt.show()
```

```python
survival_counts = df['Survived'].value_counts()
labels = [0, 1]
colors = ['lightcoral', 'lightgreen']

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(survival_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Survival Distribution - Titanic Sample')
plt.axis('equal')
plt.show()
```
